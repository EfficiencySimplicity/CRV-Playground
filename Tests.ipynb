```python
from classes.Vectorizer import *
from classes.Corpus import *
```

```python
corpus = Corpus(
    # path to the json file
    filepath = 'corpora/recipes',
    # path to the spelling dictionary
    spell_correct = 'corpora/spelling_dictionary',
    # do not seperate (e.g., bringing => bring ing)
    lemmatize = False,
    # remove any words that appear only once; replace with <UNK> in the source text
    removal_threshold = 1)
```

```python
corpus.find('kraft')
```

```python
print(random.choice(corpus.vocab))
```

```python
print(concat_sentence(random.choice(corpus.sentences)))
```

```python
# remove any words that appear 2 or less times from the CRV encodings (not keys), as the vectorizer
# uses a large matrix for efficiency, instead of dicts, which are more compact
vectorizer = corpus.create_vectorizer(removal_threshold = 2)
```

```python
# print the CRV encoding of 'half'
print(vectorizer.to_CRV('half'))
# show words similar to 'half'; the word 'half' itself can be ignored.
plot_dict(vectorizer.rate_words('half'))
```

```python
# seems 'half' appears next to itself often... how so?
corpus.find(('half', 'half'))
# ahh... makes sense now!
```

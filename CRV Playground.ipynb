```python
# lemmatization is not recommended; it adds more meanings to a word,
# thus making it harder to understand. bake has a different meaning than 'baking',
# and stuffing them all into one word leads to mild confusion.

# Note: if you run every cell in this notebook, the corpus will be replaced with a new one
# and the cells between this one and the new corpus one will no longer work,
# as they are intended for use with recipes.

corpus = Corpus(
    filepath      = 'corpora/recipes',
    spell_correct = 'corpora/spelling_dictionary',
    lemmatize     = False,
    removal_threshold = 1)

signatures = corpus.create_signatures(2)
vectorizer = corpus.create_vectorizer(removal_threshold = 2)
# vectorizer.normalize()
```

```python
CRV.PRINT_CUTOFF_AMT = 80
print(vectorizer.to_CRV('chicken') - vectorizer.to_CRV('beef'))
```

```python
corpus.find(('beef', '-'))
```

```python
# Testing different comparison methods
#
# taking the minimum of 2 CRVs will be 1 if they are equal,
# and less than 1 if there is any difference.
# since CRVs always sum to 1,
# any increase in one value equals decrease in one or more others.
#
# Taking the difference often has the same ordering of results,
# but is tricky to normalize

# Some words to test:

# celsius qt mother butter nahum sliced pasta www 
# after stir of melt you want overcook mixture let degrees
crv = vectorizer.to_CRV('mother')

plot_dict(vectorizer.rate_words(crv, mode = 'min'))
plot_dict(vectorizer.rate_words(crv, mode = 'sqrt'))
plot_dict(vectorizer.rate_words(crv, mode = 'diff'))

# corpus.find('ichimi')
```

```python
# let's look at the word 'pasta', and see what words appear near it the most
# this will have the effect so that words will have higher ratings
# the higher their value for 'pasta' is.
# Do not confuse this with the CRV for 'pasta'.

anti_CRV = CRV({'pasta' : 1})
plot_dict(vectorizer.rate_words(anti_CRV))

# Some surprises: 'seashell', 'rainbow', 'e'? 'hair' is from 'angel hair', and it wouldn't appear
# anywhere else in the recipe corpus -- a lot of pasta shapes and brands are there too.
```

```python
corpus.find(['pasta', 'seashell'],2)
corpus.find(['pasta', 'rainbow'],2)
corpus.find(['pasta', 'e'],2)
```

```python
# It appears 'seashell' is a type of pasta;
# 'rainbow pasta salad' is a recipe title;
# and pasta e <UNK> is probably some French word that was removed due to rarity.
```

```python
# Averaging together abbreviations for amounts should result in more, which it does
 
amt_CRV = vectorizer.to_CRV(vectorizer.average('qt', 'lb', 'tsp', 'c', 'tbsp'))
print(amt_CRV)
plot_dict(vectorizer.rate_words(amt_CRV, mode = 'min'))
```

```python
# Let's look at the average of oven and grill and find similar words

grill_vec = vectorizer.to_vector('grill')
oven_vec = vectorizer.to_vector('oven')
average_vec = (grill_vec + oven_vec) / 2


rated_words = vectorizer.rate_words(average_vec, 'min')
plot_dict(rated_words)

# cooking terms!
```

```python
# Let's look at the average of oven and grill 

grill_vec = vectorizer.to_vector('grill')
oven_vec = vectorizer.to_vector('oven')
average_vec = (grill_vec + oven_vec) / 2

average_CRV = vectorizer.vectorize(average_vec, mode = 'CRV')

print(average_CRV)
average_CRV.plot()

# numbers for degrees, and cooking words
```

```python
# what is the difference between 'hi' and 'greetings'?
# a positive difference means 'hi' appears next to that word more, and vice versa
print(vectorizer.to_CRV('hi') - vectorizer.to_CRV('greetings'))

# It looks like 'greetings' more often has:
# '!', 
# line breaks after, 
# 'from', as in 'greetings from' , 
# and 'my', ('greetings, my friend)

# and 'hi' more often has:
# 'hi there!', as opposed to 'greetings there!' 
# a period after,
# most countries,
# and '<Start>', (meaning most sentences start with 'hi' as opposed to 'greetings')

# That one's interesting; it sort of gives a sense as to how common a word is in a position,
# Although now that I thing about it, that's what all CRVs do.
```
